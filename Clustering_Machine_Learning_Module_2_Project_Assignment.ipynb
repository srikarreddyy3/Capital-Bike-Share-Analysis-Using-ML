###Introduction:
Bike-sharing systems are a modern evolution of traditional bike rental services,, wherein the whole procedure, encompassing membership, bike rental, and return, has been mechanized. These systems enable users to conveniently lease a bicycle at one station and return it to a different station via shuttle services. Today, there exist more than 500 bike-sharing initiatives worldwide, encompassing a fleet of over 500,000 bicycles. The increasing prevalence of these systems can be attributed basically to their substantial contributions in addressing traffic, environmental, and public health issues.

Furthermore, apart from their pragmatic uses, the data produced by bike-sharing systems offers significant prospects for scholarship. In contrast to other modes of public transportation like buses or subways, these systems offer accurate documentation of journey length, origin and destination points. This technology converts bike-sharing networks into a virtual sensor system, allowing for the observation of urban mobility patterns. Hence, it is feasible to identify significant urban occurrences by examining these observational trends.

```python
#Calculate Q1 (25th percentile) and Q3 (75th percentile) for 'duration' column
Q1 = df['duration'].quantile(0.25)
Q3 = df['duration'].quantile(0.75)
IQR = Q3 - Q1  # Interquartile Range

# Define the upper and lower bound for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Calculate the median of the 'duration' column
median_duration = df['duration'].median()

#Impute outliers (replace values outside the bounds with the median)
df['duration'] = df['duration'].apply(lambda x: median_duration if x < lower_bound or x > upper_bound else x)

# Verify if the outliers are imputed
print(f"Dataset size remains the same: {df.shape[0]}")
```

```python
#import necessary Libraries
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

from sklearn.metrics.pairwise import cosine_distances
from scipy.spatial.distance import squareform
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from sklearn.decomposition import PCA
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#  Remove all non-numeric columns and the label 'rideable_type'
# We will drop all object columns that are not useful for clustering
labels = df['rideable_type']
data = df.drop(columns=['rideable_type', 'ride_id', 'started_at', 'ended_at',
                        'start_station_name', 'end_station_name', 'start_station_id',
                        'end_station_id'])

# Convert 'member_casual' to numeric using one-hot encoding
data = pd.get_dummies(data, columns=['member_casual'], drop_first=True)

# Ensure all remaining columns are numeric
# Select only numeric columns
numeric_cols = data.select_dtypes(include=['number']).columns
data = data[numeric_cols]

#  Scale the numeric data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

#  Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(scaled_data, labels, test_size=0.3, random_state=42)

print("Scaled Data:\n", X_train[:5])
print("\nLabels:\n", y_train[:5])






```

```python
# After scaling
scaled_df = pd.DataFrame(scaled_data, columns=numeric_cols)

print("Normalized Data (before PCA):")
print(scaled_df.head())

# Save the normalized preview as an image (optional)
plt.figure(figsize=(10, 5))
plt.table(cellText=scaled_df.head().values, colLabels=scaled_df.columns, loc='center', cellLoc='center')
plt.axis('off')
plt.savefig('normalized_data.png')
```

```python

# Initialize PCA to reduce data to 3 dimensions
pca = PCA(n_components=3)

#  Fit PCA on the scaled data
pca_data = pca.fit_transform(scaled_data)

```

```python
# Calculate the explained variance ratio (percentage of variance retained)
explained_variance = pca.explained_variance_ratio_
```

```python
# Calculate the total variance retained in 3 dimensions
total_variance_retained = sum(explained_variance) * 100

print(f"Explained Variance by each component: {explained_variance}")
print(f"Total variance retained in 3D: {total_variance_retained:.2f}%")
```

```python

# Create a DataFrame from PCA-reduced data
pca_df = pd.DataFrame(pca_data, columns=[f'PC{i+1}' for i in range(3)])

print("PCA Reduced Data (3 Dimensions):")
print(pca_df.head())

# Save the PCA reduced preview as an image (optional)
plt.figure(figsize=(10, 5))
plt.table(cellText=pca_df.head().values, colLabels=pca_df.columns, loc='center', cellLoc='center')
plt.axis('off')
plt.savefig('pca_reduced_data.png')
```

```python
#since the dataset is of large size, consider only 5000 values for clustering
pca_sample_data = pca_data if pca_data.shape[0] <= 5000 else pca_data[np.random.choice(pca_data.shape[0], 5000, replace=False)]
```

```python
# Define a range for the number of clusters (k)
range_k = [2, 3, 4, 5, 6]
silhouette_scores = []

# Step 1: Ensure silhouette scores are calculated for all values of k
for k in range_k:
    try:
        kmeans = KMeans(n_clusters=k, random_state=42, max_iter=200, n_init=10)  # Removed n_jobs
        cluster_labels = kmeans.fit_predict(pca_sample_data)

        silhouette_avg = silhouette_score(pca_sample_data, cluster_labels)
        silhouette_scores.append(silhouette_avg)
        print(f"For k={k}, the average silhouette score is {silhouette_avg:.4f}")

    except Exception as e:
        print(f"Error for k={k}: {e}")

# Step 2: Check if silhouette scores were calculated properly
print(f"Silhouette scores: {silhouette_scores}")

# Step 3: Ensure that silhouette_scores has enough values to select top 3
if len(silhouette_scores) < 3:
    print("Not enough silhouette scores to select top 3 values.")
else:
    # Step 4: Sort silhouette scores in descending order and get the top 3 k values
    sorted_indices = np.argsort(silhouette_scores)[::-1]  # Sort in descending order
    top_k_values = sorted_indices[:3]  # Get the indices of the top 3 silhouette scores

    # Display the top 3 k values
    print(f"Top 3 k values based on silhouette scores: {[range_k[i] for i in top_k_values]}")



```

```python


plt.figure(figsize=(10, 6))

# Use range_k for x-axis (number of clusters) and silhouette_scores for y-axis
plt.plot(range_k, silhouette_scores, marker='o', linestyle='-', color='blue')

plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different k Values')
plt.xticks(range_k)  # Set x-axis ticks to the values of k
plt.grid(True)
plt.show()
```

**Observations:** The results of the Principal Component Analysis (PCA) clustering, evaluated through silhouette scores, indicate that k=2 yields the most distinct and well-defined clustering for the dataset, achieving the highest silhouette score of 0.4869. This suggests that the data is optimally characterized by two clusters, wherein the points within each cluster exhibit high similarity, and the distinction between the clusters is pronounced.

As the number of clusters increases to k=3, the silhouette score decreases to 0.2885, indicating a reduction in cluster clarity and an increase in overlap among groups. Despite an increase to k=4 yielding a marginal enhancement with a score of 0.3033, it remains inferior to the distinctiveness achieved at k=2.
Increasing the clusters to k = 5 and k=6 results in a decrease in silhouette scores to 0.2786 and 0.2929, respectively, indicating that additional clusters produce indistinct and overlapping groups.

In conclusion, k=2 yields the optimal clustering outcome, facilitating the most pronounced differentiation of the data into discrete groups. Although k=4 exhibits potential, k=2 is the optimal selection according to the silhouette scores, rendering it the most appropriate for this dataset.

```python
#  Compute the cosine distance matrix
cosine_dist = cosine_distances(pca_sample_data)

#Since linkage expects a condensed distance matrix, we convert the square matrix to a condensed form
condensed_cosine_dist = squareform(cosine_dist, force='tovector')

#  Perform hierarchical clustering using cosine distance
linked = linkage(condensed_cosine_dist, method='ward')

plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("Hierarchical Clustering Dendrogram (Cosine Distance)")
plt.xlabel("Data Points")
plt.ylabel("Cosine Distance")
plt.show()
```

**Observations:** The dendrogram clusters data points by orientation rather than absolute value using cosine distance. Clusters merge more gradually and uniformly at lower cosine distances in the hierarchical structure. At lower levels, branches are tighter and more uniform because most data points have similar directions, and the largest separation is around 50 cosines.

##DBSCAN clustering

```python
# Perform DBSCAN clustering on the 3D PCA-reduced data
# eps is the maximum distance between two samples for them to be considered as in the same neighborhood
# min_samples is the number of samples in a neighborhood for a point to be considered a core point
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(pca_sample_data)
```

```python
# Unique labels in DBSCAN, -1 represents noise
unique_labels = np.unique(dbscan_labels)
n_clusters = len(unique_labels) - (1 if -1 in dbscan_labels else 0)  # Subtract 1 if there is noise

print(f"Estimated number of clusters: {n_clusters}")
print(f"Number of noise points: {list(dbscan_labels).count(-1)}")
```

**Observations**The dataset had 12 clusters identified by DBSCAN. DBSCAN doesn't need a set number of clusters like KMeans. Instead, it divides points by density into dense clusters. This lets the algorithm find a natural number of clusters based on data distribution.

In addition to clusters, DBSCAN found 318 noise points, or data points without a cluster. Due to their low density and distance from other points, these points are outliers. Noise points reveal dataset anomalies.

DBSCAN output shows 12 clusters in high-density areas and 318 outliers as noise. This clarifies cluster structure and outliers.
